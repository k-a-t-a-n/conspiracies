```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px

df = pd.read_csv('data.csv')
df
```

```python
#getting columns header

colonnes = list(df.columns.values)
print(colonnes)
```

```python
#creating a dataframe from questions to change numbers in readable answers

questions = ['Q1', 'Q2', 'Q3', 'Q4', 'Q5', 'Q6', 'Q7', 'Q8', 'Q9', 'Q10', 'Q11', 'Q12', 'Q13', 'Q14', 'Q15']
dfq = df.drop(columns=[col for col in df if col not in questions])
dfq
```

```python
#replacing values

dfq.replace(1, "strongly disagree", inplace = True)

dfq.replace(2, "disagree", inplace = True)

dfq.replace(3,"neutral", inplace = True)

dfq.replace(4,"agree", inplace = True)

dfq.replace(5,"strongly agree", inplace = True)
dfq
```

```python
#selecting values for relevant columns

df_relevant = df[["education", 
                  "urban", 
                  "gender",
                  "age",
                  "hand",
                  "religion",
                  "orientation",
                  "race",
                  "voted",
                  "married",
                  "familysize",
                  "major"]]
df_relevant
```

```python
#replacing values for relevant columns


df_relevant.replace({'education' : { 0: "no answer", 1 : "Less than high school", 2 : "High school", 3 : "University degree", 4 : 'Graduate degree' }}, inplace = True)

df_relevant.replace({'urban' : { 0 : "no answer", 1 : "Rural", 2 : "Suburban", 3 : 'Urban' }}, inplace = True)

df_relevant.replace({'gender' : {0: "no answer", 1 : "male", 2 : "female", 3 : "other"}}, inplace = True)

df_relevant.replace({'hand' : { 0: "no answer", 1 : "right", 2 : "left", 3 : "both"}}, inplace = True)

df_relevant.replace({'religion' : { 0: "no answer", 1 : "Agnostic", 2:"Atheist", 3:"Buddhist", 4:"Christian (Catholic)", 5:"Christian (Mormon)", 6:"Christian (Protestant)", 7:"Christian (Other)", 8:"Hindu", 9:"Jewish", 10:"Muslim", 11:"Sikh", 12:"Other"}}, inplace = True)

df_relevant.replace({'orientation' : { 0: "no answer",1:"Heterosexual", 2:"Bisexual", 3:"Homosexual", 4:"Asexual", 5:"Other"}}, inplace = True)

df_relevant.replace({'voted' : { 0: "no answer", 1:"Yes", 2:"No"}}, inplace = True)
                                
df_relevant.replace({'married' : {0: "no answer", 1:"Never married", 2:"Currently married", 3:"Previously married"}}, inplace = True) 


df_relevant
```

```python
#adding relevant columns to questions

df2 = pd.concat([dfq, df_relevant], axis=1)
df2
```

```python
#look for correlations

df_cor = df.drop(['E1', 'E2', 'E3', 'E4', 'E5', 'E6', 'E7', 'E8', 'E9', 'E10', 'E11', 'E12', 'E13', 'E14', 'E15', 'introelapse', 'testelapse', 'surveyelapse', 'TIPI1', 'TIPI2', 'TIPI3', 'TIPI4', 'TIPI5', 'TIPI6', 'TIPI7', 'TIPI8', 'TIPI9', 'TIPI10', 'VCL1', 'VCL2', 'VCL3', 'VCL4', 'VCL5', 'VCL6', 'VCL7', 'VCL8', 'VCL9', 'VCL10', 'VCL11', 'VCL12', 'VCL13', 'VCL14', 'VCL15', 'VCL16','engnat','race'], axis=1)

plt.figure(figsize=(15, 7))
sns.heatmap(df_cor.corr(),cmap="vlag", center= 0)
plt.title('Possible correlations')

plt.show()
```

```python
#giving points for the two possible results

points = df["Q1"] + df['Q2'] + df['Q3'] + df['Q4'] + df['Q5'] + df['Q6'] + df['Q7'] + df['Q8'] + df['Q9']+ df['Q10'] + df['Q11']+ df['Q12'] + df['Q13'] + df['Q14'] + df['Q15']
df["points"] = points
df
```

```python
#adding the last column with the result based on points

df['result'] = np.where(df['points']>= 40, "complotist", "Not complotist")
```

```python
df['result'].value_counts()
```

```python
#cleaning for knn model

df_ml = df[['urban',
            'gender',
            'age',
            'hand',
            'voted',
            'married',
            'familysize',
            'points',
            'result']]
df_ml
```

```python
df_ml.to_csv(r'D:\ecole\conspiracy\df_ml.csv')
```

```python
#train test split

y = df_ml['result']
X = df_ml[['urban',
            'gender',
            'age',
            'hand',
            'voted',
            'married',
            'familysize'
            ]]
X
```

```python
#initialisating the model

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=36, train_size = 0.75)
print("The length of the initial dataset is :", len(X))
print("The length of the train dataset is   :", len(X_train))
print("The length of the test dataset is    :", len(X_test))
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()
model.fit(X, y)
print(model)
```

```python
#testing model with my random data

my_urban = 3
my_gender = 1
my_age = 54
my_hand = 3 
my_voted = 1
my_married = 2
my_familysize = 2
my_data = np.array([my_urban, my_gender, my_age, my_hand,my_voted,my_married,my_familysize]).reshape(1,7)
print(model.predict(my_data))

for i, j in zip(model.classes_, model.predict_proba(my_data)[0]):
  print("Prediction probability for:", i, "is", j)
```

```python
#evaluating model

print(model.score(X_train, y_train))
print(model.score(X_test, y_test))
from sklearn.metrics import confusion_matrix
pd.DataFrame(data = confusion_matrix(y_true = y_test, y_pred = model.predict(X_test)),
             index = model.classes_ + " ACTUAL",
             columns = model.classes_ + " PREDICTED")
```
